```python
def poverty_urban_plot(year, interval):
    income_factor = ssp1_human['3-HD-IncomeFactor'][ssp1_human['3-HD-IncomeFactor']['Year']==year]['SSP1'].values[0]
    df = slum_human['HumanInputs200812-Final5_merge'].merge(ssp1_human['1-HD-Population'], on = ['X','Y'])
    df.loc[:,'income_pp'] = df.loc[:,'hh_income'] / df.loc[:,'hh_size'] * income_factor
    
    df.loc[:,'units'] = df.loc[:,'population2015'] / df.loc[:,'hh_size']
    df.loc[:,'units'].replace([np.inf, -np.inf], np.nan, inplace=True)
    df.loc[:,'units'] = df.loc[:,'units'].fillna(0)
    
    df.loc[:,'units_size'] = df.loc[:,year] 
    
    df2 = df[df['units']>0][df[df['units']>0]['hh_size']>0].copy()
    
    df2.loc[:,'pop_hh'] = (1-df2.loc[:,'slum_pct']) * df2.loc[:,'units_size'] 
    df2.loc[:,'pop_slum'] = df2.loc[:,'slum_pct'] * df2.loc[:,'units_size'] 
    
    df2.loc[:,'income_pp_slum'] =  df2.loc[:,'income_pp'] * 0.288614 

    df2['geometry']= df2.apply(lambda row: Point(row["X"], row["Y"]), axis=1)
    p_gpd = gpd.GeoDataFrame(df2, crs=Pune_shp.crs, geometry="geometry")

    gpd_pune = gpd.sjoin(p_gpd, Pune_shp[['geometry']], op="within")


    p_gpd1 = p_gpd[~p_gpd.index.isin(gpd_pune.index)]
    urban_gpd = gpd_pune #pd.concat([gpd_pune, gpd_solapur], ignore_index=True)


    import math
    def rounddown(x):
        return int(math.floor(x / 100.0)) * 100
    def roundup(x):
        return int(math.ceil(x / 100.0)) * 100

    start_urbanhh = rounddown(urban_gpd.income_pp.min()) # Change this to your desired starting value
    end_urbanhh = roundup(urban_gpd.income_pp.max())  # Change this to your desired ending value
    
    start_slum = rounddown(urban_gpd.income_pp_slum.min()) # Change this to your desired starting value
    end_slum = roundup(urban_gpd.income_pp_slum.max())  # Change this to your desired ending value
    
    ranges = [(x, x + interval) for x in range(start_slum, end_urbanhh, interval)]

    # Initialize an empty frequency table
    frequency_table_urbanhh = {}
    frequency_table_urbanslum = {}
    
    # Populate the frequency table using DataFrame operations
    for range_ in ranges:
        # Count the number of values in the specified range
        frequency_table_urbanhh[range_[1]] = int(urban_gpd[(urban_gpd['income_pp'] >= range_[0]) & (urban_gpd['income_pp'] < range_[1])]['pop_hh'].sum())
        frequency_table_urbanslum[range_[1]] = int(urban_gpd[(urban_gpd['income_pp_slum'] >= range_[0]) & (urban_gpd['income_pp_slum'] < range_[1])]['pop_slum'].sum())   
        
    frequency_dff_urbanhh = pd.DataFrame.from_dict(frequency_table_urbanhh, orient='index',columns=['Frequency']).reset_index()
    frequency_dff_urbanslum = pd.DataFrame.from_dict(frequency_table_urbanslum, orient='index',columns=['Frequency']).reset_index()
    
    return frequency_dff_urbanhh, frequency_dff_urbanslum
```

### Strategic combination of interventions

```python
sim = ['Business\n-as-usual','Price \nregulation','Infrastructure \nmanagement',
       'Agricultural-\nto-urban','Enhanced \npiped supply','Comprehensive \naction']
sim2=np.repeat(sim,4)
sim_synergy = ['scenario1_supplyprice',
                 'scenario1_supplyphys',
                 'scenario1_agsupplyprice',
                 'scenario1_agsupplyphys',]
sim_synergy2 = ['scenario1_agsupply',
               'scenario1_agphys',
               'scenario1_agprice',
               'scenario1_agdemand',
               'scenario1_supplydemand']
```

```python
def stepwise_approach(vul, fin, fin2):
    m=2 # extreme climate change, high urbanization
    year = range(0,4,1)
    ind=0.23
    wid=0.2

    # combined 
    ## ag
    policy = 3
    d_ag = dict(zip(vul.iloc[:,1+m*3+policy*3*len(year):4+m*3+policy*3*len(year)].columns,vul.iloc[:,1+m*3:4+m*3].columns))
    df_ag2 =vul.iloc[:,1+m*3:4+m*3].subtract(vul.iloc[:,1+m*3+policy*3*len(year):4+m*3+policy*3*len(year)].rename(columns=d_ag))

    ## physical+ag
    waterprice = 1
    d_water = dict(zip(fin2.iloc[:,1+waterprice*3+m*3*len(sim_synergy2):4+waterprice*3+m*3*len(sim_synergy2)].columns,vul.iloc[:,1+m*3:4+m*3].columns))
    df_water2 =vul.iloc[:,1+m*3:4+m*3].subtract(fin2.iloc[:,1+waterprice*3+m*3*len(sim_synergy2):4+waterprice*3+m*3*len(sim_synergy2)].rename(columns=d_water))

    ## ag+physical+supply
    waterprice = 3
    d_water = dict(zip(fin.iloc[:,1+waterprice*3+m*3*len(sim_synergy):4+waterprice*3+m*3*len(sim_synergy)].columns,vul.iloc[:,1+m*3:4+m*3].columns))
    df_agdemand2 =vul.iloc[:,1+m*3:4+m*3].subtract(fin.iloc[:,1+waterprice*3+m*3*len(sim_synergy):4+waterprice*3+m*3*len(sim_synergy)].rename(columns=d_water))

    ## comprehensive
    policy = 5
    d_co = dict(zip(vul.iloc[:,1+m*3+policy*3*len(year):4+m*3+policy*3*len(year)].columns,vul.iloc[:,1+m*3:4+m*3].columns))
    df_co2 =vul.iloc[:,1+m*3:4+m*3].subtract(vul.iloc[:,1+m*3+policy*3*len(year):4+m*3+policy*3*len(year)].rename(columns=d_co))

    ## phs
    policy = 2
    d_phys = dict(zip(vul.iloc[:,1+m*3+policy*3*len(year):4+m*3+policy*3*len(year)].columns,vul.iloc[:,1+m*3:4+m*3].columns))
    df_phys2 =vul.iloc[:,1+m*3:4+m*3].subtract(vul.iloc[:,1+m*3+policy*3*len(year):4+m*3+policy*3*len(year)].rename(columns=d_phys))

    ## price
    policy = 1
    d_price = dict(zip(vul.iloc[:,1+m*3+policy*3*len(year):4+m*3+policy*3*len(year)].columns,vul.iloc[:,1+m*3:4+m*3].columns))
    df_price2 =vul.iloc[:,1+m*3:4+m*3].subtract(vul.iloc[:,1+m*3+policy*3*len(year):4+m*3+policy*3*len(year)].rename(columns=d_price))

    ## supply
    policy = 4
    d_supply = dict(zip(vul.iloc[:,1+m*3+policy*3*len(year):4+m*3+policy*3*len(year)].columns,vul.iloc[:,1+m*3:4+m*3].columns))
    df_supply2 =vul.iloc[:,1+m*3:4+m*3].subtract(vul.iloc[:,1+m*3+policy*3*len(year):4+m*3+policy*3*len(year)].rename(columns=d_supply))

    return df_ag2,df_water2,df_agdemand2,df_co2,df_phys2,df_price2,df_supply2
```
