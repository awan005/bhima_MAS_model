```python
basepath = r"/scratch/users/awan005/bhima_mas_code_publication"
path = basepath + r"/outputs/data_processed/resource_status_metrics" +"/"
res_tt2 = pd.read_csv(path + r"res_forpaper_extremeclimate_ssp1.csv").iloc[:,1:].to_dict()
river_tt2 = pd.read_csv(path + r"river_paper_extremeclimate_ssp1.csv").iloc[:,1:]
pp_bb = pd.read_csv(path + r"gw_forpaper_pune_extremeclimate_ssp1.csv").iloc[:,1].to_list()
pp_tanker = pd.read_csv(path + r"gw_forpaper_tanker_extremeclimate_ssp1.csv").iloc[:,1].to_list()
```

```python
Months = ['Jun1','July1','Aug1','Sep1','Oct1','Nov1','Dec1','Jan1','Feb1','Mar1','Apr1','May1',
         'Jun2','July2','Aug2','Sep2','Oct2','Nov2','Dec2','Jan2','Feb2','Mar2','Apr2','May2',
         'Jun3','July3','Aug3','Sep3','Oct3','Nov3','Dec3','Jan3','Feb3','Mar3','Apr3','May3',
         'Jun4','July4','Aug4','Sep4','Oct4','Nov4','Dec4','Jan4','Feb4','Mar4','Apr4','May4',
         'Jun5','July5','Aug5','Sep5','Oct5','Nov5','Dec5','Jan5','Feb5','Mar5','Apr5','May5',]
```

```python
fig = plt.figure(figsize=(7,2.5), dpi=600)

ax1 = fig.add_axes([0.1, 0.15, 0.25, 0.6])   
ax2 = fig.add_axes([0.41, 0.15, 0.25, 0.6])   
ax3 = fig.add_axes([0.745, 0.15, 0.25, 0.6])   
num=['(a)','(b)','(c)']
k=1
############################################axis 1#######################################################
ax1.plot(Months[:-12], [a+b+c+d for a,b,c,d in zip(list(res_tt2[List[4]].values())[48*k:48*k+48],
                                                     list(res_tt2[List[5]].values())[48*k:48*k+48],
                                                     list(res_tt2[List[6]].values())[48*k:48*k+48],
                                                     list(res_tt2[List[7]].values())[48*k:48*k+48])],
           label='Khadakwasla \n Complex')
ax1.axvline(x = 11.7, color = 'grey', linestyle = '--') 
ax1.axvline(x = 11.5+12, color = 'grey', linestyle = '--')
ax1.axvline(x = 11.5+12*2-0.2, color = 'grey', linestyle = '--')
ax1.set_title('Reservoir storage', fontsize=11,pad=5) 
for count in range(4):
    ax1.plot(Months[:-12], list(res_tt2[List[count]].values())[48*k:48*k+48], label=List[count])
    ax1.set_ylabel('Mm$^3$',rotation='horizontal', labelpad=10)  
    ax1.legend(fontsize=8)#,loc='center left', bbox_to_anchor=(1.0, 0.5))
    
    ax1.set_xticks(np.arange(5,len(Months[:-12]),12))
    ax1.set_xticklabels(['Avg','Yr1','Yr2','Yr3']) 
    ax1.tick_params(axis='x', which='major', pad=5) 
    
ax1.grid(axis='y')

ax1.set_yticks(np.arange(0,600.1,step=300))
ax1.set_ylim([-20,620])
ax1.spines['right'].set_visible(False)
ax1.spines['top'].set_visible(False)

############################################axis 2#######################################################
k=1
ax2.plot(Months[:-12], pp_bb[48*k:48*k+48], label='Pune Agglomeration')
ax2.plot(Months[:-12], pp_tanker[48*k:48*k+48], label='Pune Agglomeration\nand peri-urban area',color='grey')
 
ax2.set_title('Groundwater depth', fontsize=11,pad=5) 
ax2.set_ylabel('m',rotation='horizontal', labelpad=7)  
ax2.set_yticks(np.arange(0,30.1,step=10))
ax2.set_ylim([0,47])

ax2.grid(axis='y') 
ax2.invert_yaxis()

ax2.spines['right'].set_visible(False)
ax2.spines['top'].set_visible(False)
ax2.legend(fontsize=8,loc='lower right')

ax2.set_xticks(np.arange(5,len(Months[:-12]),12))
ax2.set_xticklabels(['Avg','Yr1','Yr2','Yr3']) 
# ax.tick_params(top=True, labeltop=True, bottom=False, labelbottom=False,pad=10)
ax2.tick_params(axis='x', which='major', pad=5) 

############################################axis 3#######################################################
ax3.plot(range(len(river_tt2.iloc[:,0].to_list()[365*4+1:])), river_tt2.iloc[:,0].to_list()[365*4+1:],  label='Khamgoan')

ax3.set_title('River discharge',fontsize=11, pad=5) 
ax3.set_ylabel('m$^3$/s',rotation='horizontal', labelpad=5)  

ax3.legend(fontsize=8)#,loc='center left', bbox_to_anchor=(1.0, 0.5))

ax3.set_xticks(np.arange(100,len(river_tt2.iloc[:,0].to_list()[365*4+1:]),365))
ax3.set_xticklabels(['Avg','Yr1','Yr2','Yr3']) 
ax3.tick_params(axis='x', which='major', pad=5) 
    
ax3.grid(axis='y')

ax3.set_yticks(np.arange(0,900.1,step=300))
ax3.set_ylim([-20,920])
ax3.spines['right'].set_visible(False)

ax3.spines['top'].set_visible(False)

c=0
for ax in [ax1,ax2,ax3]:
    ax.text(-0.3+c*0.04, 1.1, num[c], transform=ax.transAxes)
    plt.setp(ax.get_xticklabels(), rotation=0, horizontalalignment='center')

    for item in ([ax.title, ax.xaxis.label, ax.yaxis.label] +
                 ax.get_xticklabels() + ax.get_yticklabels() ):
        item.set_fontsize(8)
    c+=1

# plt.savefig('resource_status.pdf', bbox_inches=None)
```
