```python
basepath = r"/scratch/users/awan005/bhima_mas_code_publication"
path = basepath + r"/outputs/data_processed/human_wellbeing_metrics" +"/"
use = pd.read_csv(path + r"use09182024.csv")
vul = pd.read_csv(path + r"vul09182024.csv")
dur = pd.read_csv(path + r"stress09182024.csv")
cost = pd.read_csv(path + r"cost09182024.csv")
gin = pd.read_csv(path + r"gini09182024.csv")
```

```python
fig2, ax = plt.subplots(5,6, figsize=(7,4.375),gridspec_kw={'width_ratios': [1, 1,1,1,1,1]},dpi= 600)

year = range(0,4,1)
year_intervention = range(2,4,1)
count=2
seq_scenario = [1,3,0,2]
seq_intervention = [0,1, 2,4,3,5]
high_income = '#1b9e77'    # Teal
mid_income = '#4393c3'     # Purple
low_income = '#d95f02'     # Orange
num=['(a)','(b)','(c)','(d)','(e)']

# Add legend
legend_elements2 = [
    plt.Rectangle((0,0),0.5,0.5, facecolor=low_income, label='Low-income'),
    plt.Rectangle((0,0),0.5,0.5, facecolor=mid_income, label='Middle-income'),
    plt.Rectangle((0,0),0.5,0.5, facecolor=high_income, label='High-income'),
]

for n in range(6): 
    j = seq_intervention[n]
    
    ############# Water use per capita ########################################################################################################################

    wid=0.18
    ind=0.2
    ax[0,j].bar([year[x]-ind for x in range(2,3,1)],use.iloc[11,3+count*3:4+count*3],width=wid, zorder=3, color=low_income,label='Low income')
    ax[0,j].bar([year[x] for x in range(2,3,1)],use.iloc[10,3+count*3:4+count*3],width=wid, zorder=3, color=mid_income,label='Mid incom')
    ax[0,j].bar([year[x]+ind for x in range(2,3,1)],use.iloc[9,3+count*3:4+count*3],width=wid, zorder=3, color=high_income,label='High income')
    ax[0,j].set_xlim([1.4,2.6])        
#     ax[0,j].axhline(40,color='red', linestyle='--',linewidth=1)    
    ax[0,j].set_ylim([0,120])   
    ax[0,j].grid(axis = 'y',zorder=0, color='lightgrey', linestyle='-')
    ax[0,j].set_yticks(np.arange(0,121,step=40))  
    
    
    ############# Vulnerability ########################################################################################################################

    wid=0.18
    ind=0.2
    ax[3,j].bar([year[x]-ind for x in range(2,3,1)],vul.iloc[2,3+count*3:4+count*3],width=wid, zorder=3, color=low_income,label='Low income')
    ax[3,j].bar([year[x] for x in range(2,3,1)],vul.iloc[1,3+count*3:4+count*3],width=wid, zorder=3, color=mid_income,label='Mid incom')
    ax[3,j].bar([year[x]+ind for x in range(2,3,1)],vul.iloc[0,3+count*3:4+count*3],width=wid, zorder=3, color=high_income,label='High income')
    ax[3,j].set_xlim([1.4,2.6])            
    ax[3,j].set_ylim([0,90])   
    ax[3,j].grid(axis = 'y',zorder=0, color='lightgrey', linestyle='-')
    ax[3,j].set_yticks(np.arange(0,91,step=20))  
    
    ############# Shoratage duration ########################################################################################################################

    ax[4,j].bar([year[x]-ind for x in range(2,3,1)],dur.iloc[2,3+count*3:4+count*3],width=wid, zorder=3, color=low_income,label='Low income')
    ax[4,j].bar([year[x] for x in range(2,3,1)],dur.iloc[1,3+count*3:4+count*3],width=wid, zorder=3, color=mid_income,label='Middle income')
    ax[4,j].bar([year[x]+ind for x in range(2,3,1)],dur.iloc[0,3+count*3:4+count*3],width=wid, zorder=3, color=high_income,label='High income')
    ax[4,j].set_xlim([1.4,2.6])            
    ax[4,j].set_xticks(range(2,3))
    ax[4,j].set_xticklabels(['Yr2'], rotation=90, fontsize=8) 
    ax[4,j].set_ylim([0,10])   
    ax[4,j].grid(axis = 'y',zorder=0, color='lightgrey', linestyle='-')
    ax[4,j].set_yticks(np.arange(0,11.1,step=3))
        
    ############# Cost ########################################################################################################################

    ax[2,j].bar([year[x]-ind for x in range(2,3,1)],100*(cost.iloc[2,3+count*3:4+count*3]+cost.iloc[5,3+count*3:4+count*3]+cost.iloc[8,3+count*3:4+count*3]+cost.iloc[11,3+count*3:4+count*3]),width=wid, zorder=3, color=low_income)
    ax[2,j].bar([year[x] for x in range(2,3,1)],100*(cost.iloc[1,3+count*3:4+count*3]+cost.iloc[4,3+count*3:4+count*3]+cost.iloc[7,3+count*3:4+count*3]+cost.iloc[10,3+count*3:4+count*3]),width=wid, zorder=3, color=mid_income)
    ax[2,j].bar([year[x]+ind for x in range(2,3,1)],100*(cost.iloc[0,3+count*3:4+count*3]+cost.iloc[3,3+count*3:4+count*3]+cost.iloc[6,3+count*3:4+count*3]+cost.iloc[9,3+count*3:4+count*3]),width=wid, zorder=3, color=high_income,)
    ax[2,j].set_xlim([1.4,2.6])            
#     ax[2,4].legend(handles=legend_elements2, loc='upper right',fontsize=8, frameon=False)    
    ax[2,j].set_ylim([0,20])   
    ax[2,j].grid(axis = 'y',zorder=0, color='lightgrey', linestyle='-')
    ax[2,j].set_yticks(np.arange(0,19.9,step=5))
    
    ############# Gini ########################################################################################################################

    ax[1,j].bar([year[x] for x in range(2,3,1)],gin.iloc[0,3+count*3:4+count*3],width=wid, zorder=3, color=sns.color_palette("colorblind")[1])
    ax[1,j].axhline(0.25,color='orange', linestyle='--',linewidth=1)
    ax[1,j].set_xlim([1.4,2.6])   

    ax[1,j].set_ylim([0,0.6])  
    ax[1,j].grid(axis = 'y',zorder=0, color='lightgrey', linestyle='-')
    ax[1,j].set_yticks(np.arange(0,0.61,step=0.2))

    # Add legend
    legend_elements = [
        Line2D([0], [0], color='orange', linestyle='--',linewidth=1, label='Avg year')
    ]
#     legend_elements2 = [
#         Line2D([0], [0], color='red', linestyle='--',linewidth=1, label='40 LPCD')
#     ]
    ax[1,5].legend(handles=legend_elements, loc='upper right',fontsize=8, frameon=False)
#     ax[0,1].legend(handles=legend_elements2, loc='upper right',fontsize=8, frameon=False)
    
    count+=4          

metric = ['Liters\nper capita\nper day',
          'Gini\ncoefficient\nin water\nuse',
          'Percentage\nof income\nspent on\nwater',
          'Percentage\npopulation\nreceiving\nwater\n< 40 LPCD',
          'Number of\nconsecutive\nvulnerable\nmonths',
          ]
for n in range(6): 
    for i in range(5):
        j = seq_intervention[n]
        ax[i,0].set_ylabel(metric[i],fontsize=8,rotation='horizontal', labelpad=25,y=0.1)
        ax[0,0].set_ylabel(metric[0],fontsize=8,rotation='horizontal', labelpad=25,y=0.3)
        ax[0,j].set_title(sim[n],fontsize=8,pad=5)
    
        if (i<4):
            ax[i,j].set_xticks([])
        if (j>0):
            ax[i,j].set_yticklabels([])
        
        plt.setp(ax[i,j].get_xticklabels(), rotation=0, horizontalalignment='center')
        for item in ([ax[i,j].title, ax[i,j].xaxis.label, ax[i,j].yaxis.label] +
                     ax[i,j].get_xticklabels() + ax[i,j].get_yticklabels()):
            item.set_fontsize(8)
        ax[i,j].tick_params(axis='y',  length=0)
        ax[i,j].tick_params(axis='x', color='white')
        ax[i,j].spines['right'].set_visible(False)
        ax[i,j].spines['left'].set_visible(False)
        
# Assuming you're using matplotlib, add text labels for each row
ax[0,0].text(-1, 0.5, 'Per capita\nWater use', transform=ax[0,0].transAxes, ha='right', va='center', fontsize=11)#,weight='bold')
ax[1,0].text(-1, 0.5, 'Inequality', transform=ax[1,0].transAxes, ha='right', va='center', fontsize=11)#,weight='bold')
ax[2,0].text(-1, 0.5, 'Unaffordability', transform=ax[2,0].transAxes, ha='right', va='center', fontsize=11)#,weight='bold')
ax[3,0].text(-1, 0.5, 'Vulnerability', transform=ax[3,0].transAxes, ha='right', va='center', fontsize=11)#,weight='bold')
ax[4,0].text(-1, 0.5, 'Shortage\nduration', transform=ax[4,0].transAxes, ha='right', va='center', fontsize=11)#,weight='bold')

ax[0,0].text(-2.3, 0.5, num[0], transform=ax[0,0].transAxes, ha='right')  
ax[1,0].text(-2.3, 0.5, num[1], transform=ax[1,0].transAxes, ha='right')  
ax[2,0].text(-2.3, 0.5, num[2], transform=ax[2,0].transAxes, ha='right')  
ax[3,0].text(-2.3, 0.5, num[3], transform=ax[3,0].transAxes, ha='right')  
ax[4,0].text(-2.3, 0.5, num[4], transform=ax[4,0].transAxes, ha='right')  

fig2.tight_layout(pad=0,w_pad=-3, h_pad=0)

# plt.savefig('intervention.png',  bbox_inches='tight',pad_inches=0.2)
```
