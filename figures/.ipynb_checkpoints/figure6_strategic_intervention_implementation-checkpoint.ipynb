```python
fig = plt.figure(figsize=(7,4),dpi=600)
ax2 = fig.add_axes([0.1, 0.05, 0.4, 0.4])  

m=2 # extreme climate change, high urbanization
year = range(0,4,1)
ind=0.23
wid=0.2


sim = ['Business\n-as-usual','Price \nregulation','Infrastructure \nmanagement',
       'Agricultural-\nto-urban','Enhanced \npiped supply','Comprehensive \naction']
sim2=np.repeat(sim,4)


sim_synergy = ['scenario1_agphys',
                'scenario1_agsupplyphys',]


# combined 
## ag
policy = 3
d_ag = dict(zip(vul.iloc[:,1+m*3+policy*3*len(year):4+m*3+policy*3*len(year)].columns,vul.iloc[:,1+m*3:4+m*3].columns))
df_ag =vul.iloc[:,1+m*3:4+m*3].subtract(vul.iloc[:,1+m*3+policy*3*len(year):4+m*3+policy*3*len(year)].rename(columns=d_ag))


## physical+ag
waterprice = 0
d_water = dict(zip(fin.iloc[:,1+m*3+waterprice*3*len(year):4+m*3+waterprice*3*len(year)].columns,vul.iloc[:,1+m*3:4+m*3].columns))
df_water =vul.iloc[:,1+m*3:4+m*3].subtract(fin.iloc[:,1+m*3+waterprice*3*len(year):4+m*3+waterprice*3*len(year)].rename(columns=d_water))


## ag+physical+supply
waterprice = 1
d_supplyphys = dict(zip(fin.iloc[:,1+m*3+waterprice*3*len(year):4+m*3+waterprice*3*len(year)].columns,vul.iloc[:,1+m*3:4+m*3].columns))
df_supplyphys =vul.iloc[:,1+m*3:4+m*3].subtract(fin.iloc[:,1+m*3+waterprice*3*len(year):4+m*3+waterprice*3*len(year)].rename(columns=d_supplyphys))


## comprehensive
policy = 5
d_co = dict(zip(vul.iloc[:,1+m*3+policy*3*len(year):4+m*3+policy*3*len(year)].columns,vul.iloc[:,1+m*3:4+m*3].columns))
df_co =vul.iloc[:,1+m*3:4+m*3].subtract(vul.iloc[:,1+m*3+policy*3*len(year):4+m*3+policy*3*len(year)].rename(columns=d_co))


ax2.bar([i-ind for i in range(0,4,1)],[df_ag.iloc[0][2]/df_co.iloc[0][2],
                                             df_water.iloc[0][2]/df_co.iloc[0][2],
                                             df_supplyphys.iloc[0][2]/df_co.iloc[0][2],
                                             df_co.iloc[0][2]/df_co.iloc[0][2]], zorder=0, color=g(0.6),       
                                               
                                            label='Simultaneous implementation',width=wid)

# sum up all individual ones
## phs
policy = 2
d_phys = dict(zip(vul.iloc[:,1+m*3+policy*3*len(year):4+m*3+policy*3*len(year)].columns,vul.iloc[:,1+m*3:4+m*3].columns))
df_phys =vul.iloc[:,1+m*3:4+m*3].subtract(vul.iloc[:,1+m*3+policy*3*len(year):4+m*3+policy*3*len(year)].rename(columns=d_phys))

## price
policy = 1
d_price = dict(zip(vul.iloc[:,1+m*3+policy*3*len(year):4+m*3+policy*3*len(year)].columns,vul.iloc[:,1+m*3:4+m*3].columns))
df_price =vul.iloc[:,1+m*3:4+m*3].subtract(vul.iloc[:,1+m*3+policy*3*len(year):4+m*3+policy*3*len(year)].rename(columns=d_price))

## supply
policy = 4
d_supply = dict(zip(vul.iloc[:,1+m*3+policy*3*len(year):4+m*3+policy*3*len(year)].columns,vul.iloc[:,1+m*3:4+m*3].columns))
df_supply =vul.iloc[:,1+m*3:4+m*3].subtract(vul.iloc[:,1+m*3+policy*3*len(year):4+m*3+policy*3*len(year)].rename(columns=d_supply))

df = pd.DataFrame([pd.Series([df_ag.iloc[0][2]/df_co.iloc[0][2],np.nan,np.nan,np.nan]), 
                   pd.Series([df_ag.iloc[0][2]/df_co.iloc[0][2],np.nan,df_phys.iloc[0][2]/df_co.iloc[0][2],np.nan]), 
                   pd.Series([df_ag.iloc[0][2]/df_co.iloc[0][2],df_supply.iloc[0][2]/df_co.iloc[0][2],df_phys.iloc[0][2]/df_co.iloc[0][2],np.nan]),
                   pd.Series([df_ag.iloc[0][2]/df_co.iloc[0][2],df_supply.iloc[0][2]/df_co.iloc[0][2],df_phys.iloc[0][2]/df_co.iloc[0][2],df_price.iloc[0][2]/df_co.iloc[0][2],])],
                  index=["Single", "Two", "Three","Four"])

df = df.rename(columns={0: 'Agricultural-to-urban water transfer', 
                        1: 'Enhanced piped water supply',
                        2: 'Infrastructure management', 
                        3: 'Price regulation'})

df.plot.bar(stacked=True,ax=ax2,width=wid,color=['#fb8072','#bebada',  '#8dd3c7', '#80b1d3',],zorder=3,fontsize=8)


# Get current handles and labels
handles, labels = ax2.get_legend_handles_labels()

# Define your desired legend order
desired_order = ['Simultaneous implementation',
                 'Price regulation', 
                 'Infrastructure management', 
                 'Enhanced piped water supply', 
                 'Agricultural-to-urban water transfer']

# Create a mapping of labels to their indices
label_indices = {label: i for i, label in enumerate(labels)}

# Sort handles based on desired order
ordered_handles = [handles[label_indices[label]] for label in desired_order if label in label_indices]
ordered_labels = [label for label in desired_order if label in label_indices]

# Create a new legend with the ordered items
ax2.legend(ordered_handles, ordered_labels, 
           bbox_to_anchor=(1.05, 0.8), 
           loc='upper left', 
           borderaxespad=0., 
           fontsize=8)


# ax2.legend(bbox_to_anchor=(1.05, 0.8), 
#            loc='upper left',
#            borderaxespad=0.,fontsize=8)
# ax2.set_xlabel('Combinations of intervention portfolios',rotation=0, labelpad=5)

# ax2.axvline(0.43,linestyle='--',linewidth=0.5,color='black')

plt.ylim([0,1.0])
plt.xlim([-0.5,3.25])
plt.xticks(rotation=0, ha='center')
plt.ylabel('Reduction in\nper capita\nwater use\nGini coefficient\nrelative to\ncomprehensive\naction',rotation=0,y=0.2, labelpad=30)
plt.xlabel('Strategic implementation of intervention portfolios',rotation=0, labelpad=5)

for item in ([ax2.title,ax2.xaxis.label, ax2.yaxis.label] +
             ax2.get_xticklabels() + ax2.get_yticklabels() ):
    item.set_fontsize(8)

# plt.savefig('stepwise_synergy_legend_EF.png', bbox_inches='tight')
```
